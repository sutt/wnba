# Evaluation

```python
# 1/n * SUM[ y(i)log(yhat(i)) + (1-y(i))log(1-yhat(i)) ]
```

```python
from math import log as log

def ln(x):
    y = -30 if x == 0 else log(x) 
    return float(y)

def err(y_act, y_hat):
    y = np.mean( map( lambda x: float(x[0])*ln(x[1]) +  \
                                float(1-x[0])*ln(1-x[1]) \
                    ,zip(y_act,y_hat)))
    return -y
```

```python
ybar = np.mean(df1['Label'])
print 'ybar: ', ybar

ybase= [ybar] * df1.shape[0]
print ybase[0:5], '...'
ya, yb =  np.array(df1['Label']), np.array(ybase)

print 'true mean:  ', "%0.4f" % err( df1['Label'], ybase)
print 'higher .05: ', "%0.4f" %  err(ya, map(lambda x: x + .05, yb) )
print 'lower .05:  ', "%0.4f" %  err(ya, map(lambda x: x - .05, yb) )
```

```python
ep0 = mod1.predict_proba(df_test_nna[df_ints])
print np.mean(ep0)
len(ep0)
```

```python
out1 = map(lambda x: x[1], ep0)
print len(out1), len(df_test_nna)
```

```python
np.mean(out1)
```

```python
import scipy as sp
def ef(act, pred):
    e = -sp.mean(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    return e
```

```python
y_fit = df_fit_nna['Label']
y_test = df_test_nna['Label']
m_fit = np.mean(y_fit)
m_test = np.mean(y_test)
l_fit = len(df_fit_nna)
l_test = len(df_test_nna)

print 'fit base ', ef(y_fit,[m_fit]*l_fit)
print 'test base ', ef(y_test, [m_fit]*l_test)
print 'test opt mean ', ef(y_test, [m_test]*l_test)

print 'test yact mean ', m_test
print 'test yhat mean ', np.mean(out1)

print 'test err ', ef(y_test,out1)

e_fit = map( lambda x: x[1], mod1.predict_proba(df_fit_nna[df_ints]) )
print 'fit err ', ef(y_fit,e_fit)
```
