# Error

```python
import numpy as np
from math import log as log

def ln(x):
    y = -30 if x == 0 else log(x) 
    return float(y)

def err(y_act, y_hat):
    return -np.mean( map( lambda x: x[0]*ln(x[1]) +  (1-x[0])*ln(1-x[1]) ,zip(y_act,y_hat)))    
```

```python
n = 1000
%time err(uno[0:n],([np.mean(uno)]*n))
n = 10000
%time err(uno[0:n],([np.mean(uno)]*n))
n = 100000
%time err(uno[0:n],([np.mean(uno)]*n))
#so about 12 minutes for the whole train set 2s*100*2.5
```

```python
import scipy as sp
def llfun(act, pred):
    epsilon = 1e-15
    pred = sp.maximum(epsilon, pred)
    pred = sp.minimum(1-epsilon, pred)
    ll = sum(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    ll = ll * -1.0/len(act)
    return ll
```

```python
def ef(act, pred):
    e = -sp.mean(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    return e
```

```python
n=10000000
yact, yhat = uno[0:n], [.25] * n
%time llfun(yact,yhat)
%time ef(yact,yhat)
```

# Error Reduction from Time-Series

Testing the auto-correlation should use randomizing the intervals
to see the effect of the correct mean within the interval 

sort by randn and order labels from the original index

```python
n=int(5e6)
seed = [ (i,v) for (i,v) in enumerate(np.random.randn(n))]
dtype = [('index',int),('rando',float)]
seednp = np.array(seed,dtype=dtype)
sort1 = np.sort(seednp,order='rando')
rand1 = np.array([x[0] for x in sort1])
#rand1
```

```python
rlab1 = np.array([ uno[i] for i in rand1])
rlab1
```
