### Assess Moving Average with dif starting points for intervals

```python
gran =10000
br = np.arange(0, len(uno),step = gran)
br2 = br + (.5*gran)

def sum_over_gran(interval):
    return np.mean(uno[interval :(interval+gran)])

y,y2 = map(sum_over_gran, br),map(sum_over_gran, br2)

bplt, (sub1,sub2) = plt.subplots(nrows=2,ncols=1,figsize = (17,7) )
sub1.plot(y)
sub2.plot(y2)
```

### Build a random sample from series

```python
n=int(5e6)
seed = [ (i,v) for (i,v) in enumerate(np.random.randn(n))]
dtype = [('index',int),('rando',float)]
seednp = np.array(seed,dtype=dtype)
sort1 = np.sort(seednp,order='rando')
rand1 = np.array([x[0] for x in sort1])
```

```python
rlab1 = np.array([ uno[i] for i in rand1])
rlab1
```

### Plot Randomized Series

```python
gran =10000
br = np.arange(0, len(rlab1),step = gran)
br2 = br + (.5*gran)

def sum_over_gran(interval):
    return np.mean(rlab1[interval :(interval+gran)])

y,y2 = map(sum_over_gran, br),map(sum_over_gran, br2)

bplt, (sub1,sub2) = plt.subplots(nrows=2,ncols=1,figsize = (17,7) )
sub1.plot(y)
sub2.plot(y2)
```
